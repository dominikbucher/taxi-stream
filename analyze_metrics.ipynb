```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics = pd.read_csv('data/application-metrics.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
metrics['throughput'].plot(ax=ax, figsize=(28, 12))#, xlim=(0, 30000))
plt.axhline(y=3000, color='r', linestyle='-')
plt.show()
```

```python
metrics['backoff'].plot(figsize=(28, 12))#, xlim=(0, 15000))
```

```python
metrics['timediff'].plot(figsize=(28, 12))#, xlim=(0, 1000))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
metrics[['taxiupdates']].plot(ax=ax, figsize=(28, 12))#, xlim=(0, 30000))
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```
